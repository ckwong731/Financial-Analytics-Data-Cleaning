# Winzorising all variables outside of the 1st and 99th percentile range

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('~/desktop/FNCE2408/VERSION2ANALYTICS_FOR_FINANCE_M2_GROUP3_copy.csv')
df.head()
```

```python
df.columns
```

```python
data = ['at', 'capx', 'ceq', 'che',
       'csho', 'dlc', 'dltt', 'ebit', 'ibc', 'oiadp', 'ppent', 're', 'sale',
       'txpd', 'wcap', 'xint', 'xrd', 'sich', 'prcc_f', 'BLEV',
       'MLEV', 'DEBTISS_LAGAT', 'EQISS_LAGAT', 'CAPX_LAGAT', 'RD_LAGAT ', 'Q',
       'ROA ', 'LOG_SALES', 'LOGMKTCAP ', 'ZSCORE ', 'PPENT_AT ', 'CF',
       'CASH']
for i in data:
    firstpercentile = df[i].quantile(q=0.1)
    ninetyninthpercentile= df[i].quantile(q=0.99)
    for value in df[i]:
        if value < firstpercentile:
            dic3 = {value:firstpercentile}
            df[i].replace(dic3)
        elif value > ninetyninthpercentile:
            dic4 = {value:ninetyninthpercentile}
            df[i].replace(dic4)
```

```python
df.to_csv('~/desktop/FNCE2408_Group3_M3Assignment.csv', index = False, header=True)
```
